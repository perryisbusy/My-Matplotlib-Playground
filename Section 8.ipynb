# Part 1

```python
x = np.arange(1, 10, 1)
x
```

```python
y = x ** 5
y
```

## Functional

```python
plt.plot(x, y, ls="--", color="black")
plt.xlabel("X Lab")
plt.ylabel("Y Lab")
plt.title("Title")
plt.show()
# plt.legend()
```

```python
plt.subplot(1, 2, 1) # location
plt.plot(x, y, ls="-", color="red")

plt.subplot(1, 2, 2)
plt.plot(y, x, ls="--", color="blue")

plt.show()
```

## OO

```python
fig = plt.figure()

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(x, y, color="green")
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.set_title("Set Title")

plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # [0, 1]
ax2 = fig.add_axes([0.9, 0.1, 0.4, 0.3]) # [left x, bottom y, width, height] %

ax1.plot(x, y, color="slateblue")
ax1.set_title("ax1")

ax2.plot(y, x, color="red")
ax2.set_title("ax2")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x, y1)
ax[0].set_title("ax1")

ax[1].plot(x, y2)
ax[1].set_title("ax2")

plt.tight_layout()
plt.show()
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(3, 4), dpi=100)

ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(x, y1)


ax2 = fig.add_axes([0.1, 0.5, 0.3, 0.3])
ax2.plot(x, y2)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)


ax[0].plot(x, y1, label="ax0")
ax[0].legend()
ax[0].legend(loc=(1, 1))

ax[1].plot(x, y2, label="ax1")
ax[1].legend()
ax[1].legend(loc=(0.1, 0.8))

fig.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)


ax.plot(x, y1)

ax.plot(x, y2)

ax.legend(["x vs y1", "x vs y2"])

fig.tight_layout()
```

# Part 3

## Plot Appearance

```python
x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

ax[0][0].bar(x, y, color="red", alpha=0.1)
ax[0][1].bar(x, y, color="black", alpha=0.1)
ax[1][0].bar(x, y, color="green", alpha=0.5)
ax[1][1].bar(x, y, color="yellow", alpha=0.6)


fig.tight_layout()
```

```python
x = [1,3,4,5,7,8,10,12] # experience
y1 = [6500, 9280, 12050, 13200, 16672, 21000, 23965, 29793] # data_scientists_salary
y2 = [9020, 12873, 15725, 18000, 19790, 20196, 25769, 32000] # software_engineers_salary
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

# lw(line width), ls(line style), marker, ms(markersize) alpha (transparency)
ax.plot(x, y1, color="#4A00E0", lw=2, ls="-",
        marker="o", markersize=20, markerfacecolor="yellow", markeredgewidth=3, 
        markeredgecolor="green", alpha=0.5)
ax.set_xlim([0, 10])
ax.set_ylim([0, 100_000])

plt.show()
```

# 4 Ways

```python
x = np.linspace(start=0, stop=4*np.pi, num=1000)
y = np.sin(x)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(x, y, color="red", alpha=0.5, label="[0]")
ax[0].set_title("[0]")
ax[0].legend()

ax[1].plot(y, x, color="purple", alpha=0.5)
ax[1].set_title("[1]")
ax[1].legend("1")

fig.tight_layout()
```

```python
plt.subplot(1, 2, 1) # location
plt.plot(x, y, ls="-", color="red")

plt.subplot(1, 2, 2)
plt.plot(y, x, ls="--", color="blue")

plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # [0, 1]
ax2 = fig.add_axes([0.9, 0.1, 0.4, 0.3]) # [left x, bottom y, width, height] %

ax1.plot(x, y, color="slateblue")
ax1.set_title("ax1")

ax2.plot(y, x, color="red")
ax2.set_title("ax2")
```
